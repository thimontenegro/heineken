# Conclusão
Foi realizado dois experimentos:
## 1 - Sem as variaveis temporarais:
Sem informações de tempo, todos os modelos apresentaram baixo poder preditivo, com Test R² próximo de zero.
Mesmo modelos baseados em boosting (XGBoost e CatBoost) não conseguiram capturar sinal suficiente apenas com variáveis de processo, sugerindo que as variáveis de processo por si só não são suficientes para prever a variável alvo.

## Experimento 2 - Com variáveis temporais:
Ao incluir variáveis temporais, houve uma melhora significativa no poder preditivo dos modelos. O Test R² aumentou consideravelmente, indicando que as variáveis temporais fornecem informações cruciais para a previsão da variável alvo. Modelos como XGBoost e CatBoost se beneficiaram especialmente da inclusão de variáveis temporais, apresentando os melhores desempenhos entre os modelos testados.
 Redução consistente de MAE e RMSE
